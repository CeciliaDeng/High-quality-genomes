```bash
FASTQ_FILES=/workspace/hraczw/github/hoki_genomics/007.mappingAndCalling/fastqFiles.txt
```

```bash
python << EOF

import sys, os

fastqFiles = open('$FASTQ_FILES','r')

index = 0
for line in fastqFiles:
    if (index % 2) == 0:
        R1 = line.split('\n')[0]
    else:
        R2 = line.split('\n')[0]
        id = line.split('.R2')[0]
        os.system('bsub -J mapping -n 2 -q priority -P P/536002/01 \
        -e $WDR/' + id + '_mapping.err -o $WDR/' + id + '_mapping.out \
        "bwa mem -t 2 $REF $WDR/' + R1 + ' $WDR/' + R2 + ' | samtools view -@ 2 -Sb -o $WDR/' + id + '.bam"')
    index = index + 1
        
fastqFiles.close()
    
EOF
```

```bash
python << EOF

import sys, os


f_sampleIDFile = open('$WDR/bam_fileList.txt','r')
f_statsFile = open('$WDR/mapping_stats.txt','w')

for line in f_sampleIDFile:
    bamFile = line.split('\n')[0]
    sampleID = bamFile.split('.bam')[0]
    
    f = open('$WDR/' + sampleID + '_stats.out','r')
    for line1 in f:
        if '+ 0 in total (QC-passed reads + QC-failed reads)' in line1:
            f_statsFile.write(sampleID + '\t' + line1.split('+ 0 ')[0] + '\t')
        if '+ 0 mapped' in line1:
            f_statsFile.write(line1.split('(')[1].split(':')[0] + '\t')
        if '+ 0 properly paired' in line1:
            f_statsFile.write(line1.split('(')[1].split(':')[0] + '\n')
    f.close()

f_sampleIDFile.close()
f_statsFile.close()

EOF
```
