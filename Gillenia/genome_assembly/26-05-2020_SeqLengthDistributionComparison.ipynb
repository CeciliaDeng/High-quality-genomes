# Length distribution comparison among contig sets

```r
library(ggplot2)
```

```r
library(scales)
```

```r
setwd('/workspace/hraczw/github/GA/Gillenia_genome/004.Scaffolding_10Xcontigs/k21_t20_d5000')
getwd()
```

```r
scaffold_i1 = read.table('scaffolds_k21_t20_d5000.scaffolds.fa.fai', header = F, sep = ",")
```

```r
contig = read.table('/output/genomic/plant/Gillenia/trifoliata/31_ReleaseS4/01_Release/G3_2_S4.1.2KB.fasta.fai', 
                            header = F,
                            sep="\t")
```

```r
scaffold_i2 = read.table('scaffolds_k21_t10_d10000.scaffolds.fa.fai', header = F, sep = ",")
```

```r
scaffold_slr = read.table('/workspace/hraczw/github/GA/Gillenia_genome/004.Scaffolding_10Xcontigs/SLR/scaffold_set.fa.fai')
```

```r
ggplot(contig, aes(x = V2)) + 
scale_x_log10(breaks = trans_breaks("log10", function(x) 10^x),
              labels = trans_format("log10", math_format(.x))) +
geom_histogram() +
geom_hline(yintercept = 1000, linetype="dashed", color = "red") +
geom_vline(xintercept = 592835.692789, linetype="dashed", color = "blue") +
geom_text(aes(x=592835.692789, label="N50", y=500), colour="blue", angle=90, vjust = 1.2, text=element_text(size=15))+
labs(title = "Contig length distribution", x = "Log10_Length (bases)", y = "Count")
```

```r
ggplot(scaffold_i1, aes(x = V2)) + 
scale_x_log10() +
geom_histogram() +
geom_hline(yintercept = 1000, linetype="dashed", color = "red") +
geom_vline(xintercept = 1000000, linetype="dashed", color = "blue") +
#geom_density() + 
#xlim(0,100000) +
#scale_x_log10() #+
#ylim(0,0.00025) +
labs(title = "Initial scaffold length distribution", x = "Log10_length (bases)", y = "Count")
```
