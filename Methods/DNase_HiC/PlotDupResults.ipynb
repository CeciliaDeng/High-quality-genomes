```r
flagr<-read.csv("flag_results.csv")
```

```r
flagr$Percent_Duplicated<-(flagr$Duplicates/flagr$Mapped)*100
flagr$UseableReads<-flagr$Mapped-flagr$Duplicates
```

```r
head(flagr)
```

```r
library(ggplot2)
library(tidyverse)
```

```r
small <- flagr %>% filter(Mapped < 500000000)

e <- ggplot(small,aes(Reads,Percent_Duplicated,color=Org))
```

```r
e+geom_point() +scale_x_log10() + labs(x="Log Transformed # of HiC reads", y="Percent PCR duplication") + theme_dark() + scale_colour_brewer(type = "seq", palette = "Accent")
```

```r
ggsave("HiC_PCRdup_byPercent.png")
```

```r
f<- ggplot(small,aes(Reads,UseableReads,color=Org))
#f+geom_point()+scale_x_log10()+scale_y_log10()
#f+geom_point()
#f+geom_point() + geom_smooth(formula = y~log(x)) + scale_x_log10()+ scale_y_log10()
f+geom_point() + labs(x="Number of HiC reads", y="Number of HiC reads after removing PCR duplicates")+ theme_dark() + scale_colour_brewer(type = "seq", palette = "Accent")
```
